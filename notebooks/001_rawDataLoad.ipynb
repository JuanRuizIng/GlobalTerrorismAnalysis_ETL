```python
import pandas as pd
import connection

from sqlalchemy import text
```

### **Reading the dataset**

```python
df = pd.read_csv('../data/globalterrorismdb_0718dist.csv', low_memory=False)
```

```python
df
```

### **Transfering the data to the database in PostgreSQL**

```python
engine = connection.creating_engine()
```

```python
df.to_sql('global_terrorism_db_raw', engine, if_exists='replace', index=False)
```

### **Verifying the data transference**

```python
with engine.connect() as conn:
    result = conn.execute(text("SELECT * FROM candidates_raw LIMIT 15"))
    for row in result:
        print(row)
```
